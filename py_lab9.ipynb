```
import numpy as np

dtypes = [('name','S10'),('year','int'),('cgpa','float')]
values = [('dishen',2018,9.31),('darshan',2019,8.95),('dhruv',2017,9.01),('sharwil',2018,8.78)]
a = np.array(values,dtype=dtypes)
print(np.sort(a,order='name'))
print(np.sort(a,order=['cgpa','year']))
```

```
from numpy import *
a = array([1,2,3,4])
b = a
print(id(a)) # both are pointing to same array
print(id(b)) # ^
b = a.view() # create array at diff position (shallow copy)
print(id(a))
print(id(b))
b = a.copy() # deep copy
print(id(a))
print(id(b))
```

```
from functools import reduce
n = [3,2,4,7,8,9,1]
evens = filter(lambda n:n%2==0,n)
doubles = map(lambda n:n*n,evens)
sum = reduce(lambda a,b:a+b, doubles)
print(doubles)
print(sum)
```
