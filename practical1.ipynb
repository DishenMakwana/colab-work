```
N = 5
for i in range(1, N+1):
  for j in range(1, i + 1):
    print(j, end=' ') 
  print("")
```

```
n = 0
for i in range(n,5):
  for j in range(5-i,0,-1):
    print(j,end=" ")
  print("\r")
```

```
a=9
for i in range(1,a):
  for j in range(i-1,-1,-1):
    print(2**j,end=' ')
  print("\n")
```

```
s1="PyThOn"
s2=""
s3=""
for i in s1: 
  if i>= 'A' and i<= 'Z':
    s2+=i
  else:
    s3+=i
print(s2+s3)
```

```
s="PyT3Hon@P2"
l,d,sp = 0,0,0
for i in s:
  if i>='A' and i<='z':
    l+=1
  elif (i.isdigit()):
    d+=1
  else:
    sp+=1
      
print("letter-> ",l,"digit-> ",d,"special characters-> ",sp)
```
