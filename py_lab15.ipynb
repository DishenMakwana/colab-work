```
import requests 
page = requests.get("https://mausam.imd.gov.in/")
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```
today = soup.find(id="today")
forecast_items = today.find_all(class_="capital")
t = forecast_items[1]
print(t.prettify())
```

```
city = t.find('h3').get_text()
print("city :",city)
temp = t.find(class_="now").get_text()
wind = t.find(class_="wind").get_text()
min_max = t.find(class_="minmax").get_text()
print("Temp :",temp)
print("Wind :",wind)
print("Min-Max :",min_max)
```

```
page = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.XEI0plwza00")
soup = BeautifulSoup(page.content,'html.parser')
soup
```

```
seven_day = soup.find(id="seven-day-forecast-list")
seven_day
```

```
forcast_items = seven_day.find_all(class_="tombstone-container")
forecast_items
```

```
tonight = forcast_items[3]
tonight
```

```
period = tonight.find(class_="period-name").get_text()
short_desc = tonight.find(class_="short-desc").get_text()
temp = tonight.find(class_="temp").get_text()
print(period)
print(short_desc)
print(temp)
```

```
img = tonight.find('img')
desc = img['title']
alt = img['alt']
print(img)
print(desc)
print(alt)
```

```
periods = []
periods_tags = seven_day.select(".tombstone-container .period-name")
for item in periods_tags:
    periods.append(item.get_text())
```

```
periods
```

```

short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
descs = [d['title'] for d in seven_day.select(".tombstone-container img")]
```

```
temps.append("High: 59 °F")
print(short_descs)
print(temps)
print(descs)
```

```
weather = pd.DataFrame({
    "period": periods,
    "short_desc": short_descs,
    "temp":temps,
    "desc":descs
})
```
