```
l = [87,45,41,65,94,41,99,94] 
ll=[]
for i in range (0,len(l)):
  if l not in ll: 
    ll.append(l[i]) 
      
print(ll)
s = set(ll)
t = tuple(s) 
print(t)
print(max(t),' ',min(t))
```

```
a = {"name":"abc", "age":19}
print(a['name'])
print(a.get('name'))
a['name']='pqr'
print(a)
a['email']='abc@pqr.com'
print(a)
print(' ')

for x in a:
  print(x)
print(' ')

for x in a:
  print(a[x])
print(' ')

for x in a.values():
  print(x)
print(' ')

for x,y in a.items():
  print(x,y)
print(' ')

print(len(a))
print(' ')
a.pop('age')
print(a)
print(' ')

del a['email']
print(a)

del a
a = {"name":"abc", "age":19}
b = a.copy()
print(b)

a = {x:x*x*x for x in range(1, 8)}
print(a)

a = {"abc":{"name":"xyz", "age":19},"pqr":{"name":"jkl", "age":16}}
print(a)
print(a['abc']['name'])
```
