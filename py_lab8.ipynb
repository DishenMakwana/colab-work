```
f1 = open('file1.txt','w')
f1.write("\nline1\nline2\nline3\nline4\nline5\nline6\nline7\nline8")
f1.close()
f1 = open('file1.txt','r')
f = f1.readlines()
f2 = open("file2.txt", "w")
for i in range(1,len(f)):
    if (i%2==0):
        f2.write(f[i])
f2.close()
f2 = open("file2.txt", "r")
print(f2.read())
f2.close()
f1.close()
```
