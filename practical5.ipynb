```
import matplotlib.pyplot as plot
from numpy import *
#linerepresentationofgraph
v=[1,5,8,9,2,0,3,10,4,7]
v1=[3,8,9,2,1,2,4,7,6,6]
plt.plot(range(1,11),v,'--')
plt.plot(range(1,11),v1,':')
plt.show()

#colorrepresentation
v=[1,5,8,9,2,0,3,10,4,7]
v1=[3,8,9,2,1,2,4,7,6,6]
plt.plot(range(1,11),v,'r--')
plt.plot(range(1,11),v1,'m:')
plt.show()
'''
'''
#markerrepresentation
v=[1,5,8,9,2,0,3,10,4,7]
v1=[3,8,9,2,1,2,4,7,6,6]
plt.plot(range(1,11),v,'*--')
plt.plot(range(1,11),v1,'D:')
plt.show()

#labels,annotation,legends
v=[1,5,8,9,2,0,3,10,4,7]
v1=[3,8,9,2,1,2,4,7,6,6]
plt.xlabel('Entries')
plt.ylabel('Values')
plt.plot(range(1,11),v,'o--')
plt.plot(range(1,11),v1,'v:')
plt.show()

#annotationingraph
v=[1,5,8,9,2,0,3,10,4,7]
plt.annotate(xy=[1,1],s='FirstEntry')
plt.plot(range(1,11),v)
plt.show()

```

```
import matplotlib.pyplot as plt
from numpy import *
from pandas import *
#pie chart
values = [5,8,5,12,3,6,1]
colors = ['b','g','r','c','m','y','c']
explode = (0,0.2,0,0,0,0,0.5)# away from pie chart as a cut piece of cake
labels = ['A','B','C','D','E','F','G']
plt.pie(values,colors=colors,labels=labels,explode=explode,autopct='%1.1f%%',counterclock=False, shadow=True)
plt.title('Values')
plt.show()

#bar chart
values = [5,8,9,10,4,7]
colors = ['b','r','b','b','b','b']
widths = [0.5,0.9,0.5,0.5,0.5,0.5] #width of straight pipe
plt.bar(range(0,6),values,width=widths,color=colors,align='center')
plt.show()
```
