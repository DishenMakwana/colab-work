```
d = {
 "name": "ABC",
 "age":25,
 "salary": 15000,
"city": "Rajkot"
}
for i in (d.keys()):
  if(i=="city"):
    d["location"]=d.pop("city")
    
print(d)
```

```
d={'se':82,'ada':65,'py':75}
mx,mn=0,100
m,n="",""
for i,j in (d.items()):
  if(mx<j):
    mx=j
    m=i
  elif(mn>j):
    mn=j
    n=i
print(m,"->",mx,n,"->",mn)
```
