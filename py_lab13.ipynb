```
import matplotlib.pyplot as plt
from numpy import *
from pandas import *

#pie chart
values = [5,8,5,12,3,6,1]
colors = ['b','g','r','c','m','y','c']
explode = (0,0.2,0,0,0,0,0.5)# away from pie chart as a cut piece of cake
labels = ['A','B','C','D','E','F','G']

plt.pie(values,colors=colors,labels=labels,explode=explode,autopct='%1.1f%%',counterclock=False, shadow=True)
plt.title('Values')
plt.show()
```

```
import matplotlib.pyplot as plt
from numpy import *
from pandas import *

#bar chart
values = [5,8,9,10,4,7]
colors = ['b','r','b','b','b','b']
widths = [0.5,0.9,0.5,0.5,0.5,0.5] #width of straight pipe

plt.bar(range(0,6),values,width=widths,color=colors,align='center')
plt.show()
```

```
import matplotlib.pyplot as plt
from numpy import *
from pandas import *

#scatter plotting
x1=5*random.rand(50)
x2=5*random.rand(50)+25
x3=25*random.rand(25)
x = concatenate((x1,x2,x3))

y1=5*random.rand(50)
y2=5*random.rand(50)+25
y3=25*random.rand(25)
y = concatenate((y1,y2,y3))
color = ['b']*50 + ['g']*50 + ['r']*25
plt.scatter(x,y,s=[50],marker='^',color=color)
plt.show()
```

```
import matplotlib.pyplot as plt
from numpy import *
from pandas import *

weather = read_csv( "https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv" )
print(weather)

#bar plot
weather.plot.bar(y=['Tmax','Tmin'],x='Month')       # barh.() is for horizontal bar
plt.show()

#subplot
weather.plot.bar(y=['Tmax','Tmin','Rain','Sun'],x='Month',subplots=True,layout=(2,2))
plt.show()
```

```
#scatter plot

weather.plot.scatter(x='Sun',y='Rain')
plt.show()
```
