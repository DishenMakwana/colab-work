```
#2
#Name , Age , City
#Find Duplicate record and remove duplicate records.

import pandas as pd
import numpy as np

employees = [('Harsh', 20,'Varanasi'),
             ('Jugal', 21, 'Delhi'),
             ('Rashesh', 20, 'Mumbai'),
             ('Jugal', 21, 'Delhi'),
             ('Shreya', 21, 'Delhi'),
             ('Rutvik', 20, 'Mumbai'),
             ('Dishen', 21, 'Dehradun'),
             ('Shreya', 21, 'Delhi')
             ]
# remove duplicates only             
dfObj = pd.DataFrame(employees,columns=["Name","Age","City"])
print(dfObj)
print()
duplicatedRow = dfObj[dfObj.duplicated(keep='last')] #find duplicates
print(duplicatedRow)
print()
final_data = dfObj.drop_duplicates() #remove duplicates
#final_data.sort_values("Name",inplace=True)
print(final_data)
```

```
import pandas as pd
import numpy as np

employees = [('Harsh', 20,'Varanasi'),
             ('Jugal', 21, 'Delhi'),
             ('Rashesh', 20, 'Mumbai'),
             (NaN, 21, 'Delhi'),
             ('Shreya', 21, 'Delhi'),
             ('Rutvik', 20, NaN),
             ('Dishen', 21, 'Dehradun'),
             ('Shreya', NaN, 'Delhi')
             ]
             
df=DataFrame(employees,columns=["Name","Age","City"])
print(df.isnull()) # return boolean dataframe
print(df.isnull().sum()) # for perticular column
print(df.isnull().sum().sum()) # total values
```
